## Imports

```python
from datetime import datetime
import matplotlib.pyplot as plt

from ooijh import FLORT
```

## Define Reference Designators And Instantiate Data Class

```python
%%time
site = 'CE02SHSP'
node = 'SP'
bdt = datetime(2019,7,1)
edt = datetime(2019,8,1,23,59,59)

flort = FLORT(site, node, begin_datetime = bdt, end_datetime = edt)
```

## Verify Files

```python
flort.files
```

## Import Data

```python
flortds = flort.get_data()
```

# Plot Data

```python
fig, ax = plt.subplots(1,1,figsize = (11,8.5))

p = ax.scatter(flortds.time, flortds.depth, c =  flortds.fluorometric_chlorophyll_a, cmap = 'jet', vmin = 0, vmax = 10)
cbar = fig.colorbar(p, ax = ax, label = r'FCHL ($\frac{{\mu}g}{L}$)')
ax.invert_yaxis()
ax.set_ylim(80,0)
ax.set_ylabel('Depth (m)')
ax.set_xlabel('Datetime (UTC)')
```
